```python
# display the movie
def display_movie(fpath):
    """
    Display a movie file in a Jupyter notebook.
    
    Parameters
    ----------
    fpath : str
        Path to the movie file. 

    Returns
    -------
    None
        Displays the movie in the notebook.
    """
    from IPython.display import HTML
    from base64 import b64encode

    # read the movie file
    with open(fpath, "rb") as f:
        movie = f.read()

    # encode the movie in base64
    movie_encoded = b64encode(movie).decode("utf-8")

    # create HTML to display the movie
    html = f"""
    <video width="640" height="480" controls>
        <source src="data:video/mp4;base64,{movie_encoded}" type="video/mp4">
        Your browser does not support the video tag.
    </video>
    """
    
    return HTML(html)

# display the movie
display_movie(fpath_data)
    
```

```python
# display a movie as a matplotlib animation
def display_movie_matplotlib(fpath):
    """
    Display a movie file using matplotlib.

    Parameters
    ----------
    fpath : str
        Path to the movie file.
    
    Returns
    -------
    None
        Displays the movie using matplotlib.
    """
    
    # import libraries for animation
    from matplotlib.animation import FuncAnimation
    from matplotlib import animation
    import cv2

    # create a figure and axis
    fig, ax = plt.subplots()

    # read the movie file
    cap = cv2.VideoCapture(fpath)

    # get the frame count
    n_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # function to update the frame
    def update(frame):
        ret, img = cap.read()
        if ret:
            ax.clear()
            ax.imshow(img)
        return ax,

    # create the animation
    ani = FuncAnimation(fig, update, frames=n_frames, interval=50)

    plt.show()

# display the movie using matplotlib
display_movie_matplotlib(fpath_data)
```
