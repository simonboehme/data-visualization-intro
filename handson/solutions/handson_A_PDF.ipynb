# Simple example: import and display some PDF data

Learning goals:
- Load some simple x-y experimental data using the pandas library
- Plot the data using the matplotlib library

We will use here a PDF data set provided by Dr Adam F. Sapnik (Kopenhagen). Note that we here do not intend to interpret the displayed data, we really just want to generate a plot. That's it.

## Load python libraries

```python
# load numerical analysis library.
import numpy as np

# load data handling libraries.
import pandas as pd

# load plotting libraries.
import matplotlib
from matplotlib import pyplot as plt

# file handling libraries.
import os
from pathlib import Path
```

## Load experimental data

```python
# provide the path to the data file
fpath_data = "./data/PDF/Pt-nanoparticles.gr"             # this is some PDF example file provided by Dr. Adam F. Sapnik in the context of the SCS Summer School on Materials Chemistry 2025.
```

```python
# load data from file
data = pd.read_csv(fpath_data, skiprows=19, comment="#", delimiter=" ", header=None)                # here, we skip the first 25 lines, which are comments, and we specify that the data is space-separated.

# report it
data
```

## Plot

```python
# provide some plotting parameters
figsize = (8, 3)            # figure size in inches
color = "cornflowerblue"    # color for the trace
linewidth = 1               # line width for the trace
label = fpath_data.split("/")[-1].split(".gr")[0]  # this is just to get the file name from the path, which we will use as a label
xlabel = "r (Å)"            # x-axis label
ylabel = r"G (Å$^{-2}$)"     # y-axis label, using TeX formatting for the units
fig_fpath = "figure_A_PDF.png"  # file path for the figure

# plot the data
fig, ax = plt.subplots(figsize=figsize)
ax.plot(data[0], data[1], color=color, linewidth=linewidth, label=label)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.legend()

# save it to a file in the current working directory
fpath_out = os.path.join(os.getcwd(), fig_fpath)
fig.savefig(fpath_out, dpi=300, bbox_inches="tight")
```
