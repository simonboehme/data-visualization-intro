# Importing and displaying a published SAXS data set

```python
# load numerical analysis library.
import numpy as np

# load data handling libraries.
import pandas as pd

# load plotting libraries.
import matplotlib
from matplotlib import pyplot as plt

# file handling libraries.
import os
from pathlib import Path
```

```python
# path to the Excel data file
fpath_data = "./data/Wang_Nat_Commun_2025_16_2687/41467_2025_58028_MOESM3_ESM.xlsx"
# fpath_data = str(Path(r"C:\Users\bosi\Dropbox\ETH\boards\SCS_Materials_Chemistry\Symposium\MatChem25\school\data_visualization\data-visualization-intro\data\Wang_Nat_Commun_2025_16_2687\41467_2025_58028_MOESM3_ESM.xlsx"))
```

```python
# read data from the Excel file
df = pd.read_excel(fpath_data, sheet_name="Figure 1 b", header=[0, 1])
df
```

```python
# Extract every second column starting from index 1 (the y data columns)
y_data_columns = df.columns[1::2]
y_data = df[y_data_columns].to_numpy()
y_data.shape  # (941, 20)
```

```python
# Extract the x data columns
x_data_columns = df.columns[0::2]           # every second column starting from index 0
x_data = df[x_data_columns].to_numpy()
x_data.shape

# Test if all x_data columns are equal
if np.all(x_data == x_data[:, 0, None]):
    x_data = x_data[:, 0]  # use the first column as x data
else:
    raise ValueError("Not all x_data columns are equal. Please check the data.")

# Extract the y data columns 
y_data_columns = df.columns[1::2]           # every second column starting from index 1
y_data = df[y_data_columns].to_numpy()
y_data.shape

# Extract the z data from the df header
z_data_columns = df.columns[0::2]           # every second column starting from index 0
z_data = z_data_columns.get_level_values(0).to_numpy()
# Check if z_data is a 1D array
if z_data.ndim == 1:
    z_data = z_data[:, None]  # convert to 2D array with one column
else:
    raise ValueError("z_data is not a 1D array. Please check the data.")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
# Plot the data
for i in range(y_data.shape[1]):
    ax.plot(x_data, y_data[:, i], label=z_data[i, 0])
ax.set_yscale("log")  # Set y-axis to logarithmic scale
# Set the axis labels and title
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("Heat capacity (J/mol K)")
ax.set_title("Heat capacity vs Temperature")
# Add a legend
ax.legend(title="Phase", bbox_to_anchor=(1.05, 1), loc='upper left')


```
